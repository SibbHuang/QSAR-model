### Import Package:

```python
import pandas as pd
import pyqsar
from pyqsar import data_tools as dt  #widget for data
from sklearn.preprocessing import MinMaxScaler #Standardisation of data
import pickle
import numpy as np
```

### Check Model Data:

```python
f= ".\\f1_45rmc_csv\\fs9_feature table.csv"
data = pd.read_csv(f)
xdata = data.drop(["pIC50"],axis=1)
xdata.head()
```

```python
with open("datad9.pickle", "rb") as f:
    model = pickle.load(f)
xd = xdata.values
result = np.around(model.predict(xd),3)
pd.DataFrame(result).to_csv(".\pIC50report_d9.csv",index=False)
```

### Get New Xdata:

```python
#get columns name:
columns_data = xdata.columns.tolist() 
columns_data.remove('khs.dsN')
columns_data.append("ndsN")
```

```python
filename = ".\\peptideDresult.csv" #open peptides descriptors data
newdata = pd.read_csv(filename)

Pdata = pd.DataFrame(newdata ,columns=columns_data)
Pdata.head()
```

### Standardisation of Data:

```python
header = list(Pdata.columns.values)
scaler = MinMaxScaler()
P_data_scaled = scaler.fit_transform(Pdata)
Pdata = pd.DataFrame(P_data_scaled, columns=header)
Pdata.head()
```

### Predict activity of Novel compounds:

```python
#open the model and predict new data:
with open("datad12.pickle", "rb") as f:
    model = pickle.load(f)

px = Pdata.values
result = np.around(model.predict(px),3)
pd.DataFrame(result).to_csv(".\ppIC50report_d12.csv",index=False)
```
